```python
import numpy as np
from PIL import Image
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
img_path = "/home/tiendat/Documents/intelFPGA_projects/ImageProcessing_FPGA/img.jpeg"

image = Image.open(img_path)


# summarize image details
print(image.mode)
print(image.size)

plt.imshow(image)
```

```python
img_arr = np.asarray(image)

print(img_arr.shape)
```

```python
new_img_arr = np.zeros(shape=(*(img_arr.shape[0:2]), 1))

print(new_img_arr.shape)
```

```python
#convert to gray scale
for i in range(img_arr.shape[0]):
    for j in range(img_arr.shape[1]):
        new_img_arr[i][j] = sum(img_arr[i][j]) / 3

plt.imshow(new_img_arr, cmap='gray')
```

```python
red = 0.3
green = 0.59
blue = 0.11
```

```python
#convert to gray scale
for i in range(img_arr.shape[0]):
    for j in range(img_arr.shape[1]):
        new_img_arr[i][j] = img_arr[i][j][0] * red + img_arr[i][j][1] * green + img_arr[i][j][2] * blue

plt.imshow(new_img_arr, cmap='gray')
```
